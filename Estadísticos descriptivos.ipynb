```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Generar un vector de 500 datos con distribucin normal, media en cero, varianza en 1
datos = np.random.normal(loc=0, scale=1, size=500)

# 2) Graficar un histograma de los datos
plt.figure(figsize=(10, 5))
plt.hist(datos, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.title('Histograma de datos con distribucin normal')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# 3) Generar dos vectores ms con diferentes medias y varianzas
datos_media_modificada = np.random.normal(loc=2, scale=1, size=500)  # Modificar media a 2
datos_varianza_modificada = np.random.normal(loc=0, scale=2, size=500)  # Modificar varianza a 2

# 4) Mostrar las tres grficas y comentar cmo variaron
plt.figure(figsize=(15, 5))

# Histograma para datos con distribucin normal, media en cero, varianza en 1
plt.subplot(1, 3, 1)
plt.hist(datos, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.title('Media: 0, Varianza: 1')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True)

# Histograma para datos con media modificada
plt.subplot(1, 3, 2)
plt.hist(datos_media_modificada, bins=30, color='green', edgecolor='black', alpha=0.7)
plt.title('Media: 2, Varianza: 1')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True)

# Histograma para datos con varianza modificada
plt.subplot(1, 3, 3)
plt.hist(datos_varianza_modificada, bins=30, color='red', edgecolor='black', alpha=0.7)
plt.title('Media: 0, Varianza: 2')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True)

plt.tight_layout()
plt.show()
```
